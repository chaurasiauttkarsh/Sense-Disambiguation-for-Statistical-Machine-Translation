## Extract Best Disambiguated word for the given context(en)

```python
context_words_max = 5 # change this paramters to 1, 2, 5, 10, 15, 20 : it may improve the results
context_window_size = 10 # this parameters can be also changed during experiments 
word = "a" 
#word = "घर"
context = "a country considered as an organized political community controlled by one government."
#context = "यह मूक स्नेह था, खूब ठोस, रस और स्वाद से भरा हुआ।रस शब्द का उपयोग प्रेमचंद ने अपनी कहानी ईदगाह इस प्रकार किया है.कितना सुहावना गीत है! मुझे तो इसमें रस और भाव कूट-कूटकर भारा हुआ प्रतीत होता है।"
#context = "कीट-पतंग चारों तरफ रेंगते दिखायी देते थे।पतंग शब्द का उपयोग प्रेमचंद ने अपनी कहानी नाग-पूजा इस प्रकार किया है. बच्चे उस डिब्बे को लूटने के लिए कटी पतंग की तरह झपट पड़ते।पतंग शब्द का उपयोग प्रमोद भार्गव ने अपनी कहानी नक्टू इस प्रकार किया है."
#context = "मैंने उस औरत की बातों में आकर अपने सारे घर का चिट्ठा बयान कर दिया।इसके विरुद्ध रग्घू अपने घर का अकेला था, वह भी दुर्बल, अशक्त और जवानी में बूढ़ा।"
ignore_case = True
lang = "en" # to filter out stopwords

# Load models (takes long time)
sv = sensegram.SenseGram.load_word2vec_format(sense_vectors_fpath, binary=False)
wv = KeyedVectors.load_word2vec_format("/content/sensegram/model/eng_40000.txt.cbow1-size300-window5-iter5-mincount10-bigramsFalse.word_vectors", binary=False, unicode_errors="strict")


tokens = word_tokenize(context)
print(tokens)



# Play with the model (is quick)
print("Probabilities of the senses:\n{}\n\n".format(sv.get_senses(word, ignore_case=ignore_case)))


for sense_id, prob in sv.get_senses(word, ignore_case=ignore_case):
    print(sense_id)
    print("="*20)
    for rsense_id, sim in sv.wv.most_similar(sense_id):
        print("{} {:f}".format(rsense_id, sim))
    print("\n")

# Disambiguate a word in a context
wsd_model = WSD(sv, wv, window=context_window_size, ignore_case=ignore_case)    
print(wsd_model.disambiguate(context, word))


if len((sv.get_senses(word, ignore_case=ignore_case))) != 0:
  context = re.sub(""+word+" ", sv.get_senses(word, ignore_case=ignore_case)[0][0], context)
  print(context)
```

## Loop over the sentences and store in output file

```python
context_words_max = 5 # change this paramters to 1, 2, 5, 10, 15, 20 : it may improve the results
context_window_size = 10 # this parameters can be also changed during experiments 
context = "a country considered as an organized political community controlled by one government"
ignore_case = True
lang = "en" # to filter out stopwords

# Load models (takes long time)
sv = sensegram.SenseGram.load_word2vec_format(sense_vectors_fpath, binary=False)
wv = KeyedVectors.load_word2vec_format("/content/sensegram/model/eng_40000.txt.cbow1-size300-window5-iter5-mincount10-bigramsFalse.word_vectors", binary=False, unicode_errors="strict")

# Open the dataset
file1 = open('/content/drive/MyDrive/NLP/40000_en/eng_40000.txt', 'r')

# Read individual lines of text
Lines = file1.readlines()
count = 0
# Loop over all the individual lines
for line in Lines:
    # Tokenize the line
    count += 1
    if count % 500 == 0:
      print(count)
    tokens = word_tokenize(line)
    # Create an empty string to store output
    out_string = ""
    # Loop word by word in a given line
    for i in range(len(tokens)):
      # If disambiguated word exists append output string
      try:
        out_string = out_string + sv.get_senses(tokens[i], ignore_case=ignore_case)[0][0] + " "
      # else append the original word
      except:
        out_string = out_string + tokens[i] + " "
    # Open output file in append format
    file2 = open("/content/drive/MyDrive/NLP/out_en.txt", "a")
    file2.write(out_string + '\n')
file2.close()
```
